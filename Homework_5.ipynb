```python
# Task 1

# Write a function that takes a string and returns it reversed.
# Input: "Python" → Output: "nohtyP"

def reverse_str(string):
  return string[::-1]

print(reverse_str("Python"))
```

```python
# Task 2

# Write a function is_palindrome(text) that checks if a string is the same forward and backward.
# "madam" → Palindrome

def is_palindrome(text):
  if text == text[::-1]:
    print("Palindrome")
  else:
    print("Not Palindrome")
```

```python
is_palindrome("madam")
is_palindrome("SMIT")
```

```python
# Task 3

# Write a program that asks for a sentence and counts:
# Digits
# Alphabets
# Spaces
def counts():
  sentence = input("Enter a sentence: ")
  digits = sum ( 1 for i in sentence if i in "1234567890")
  alphabets = sum (1 for a in sentence if a in "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ")
  spaces = sum (1 for s in sentence if s in " ")
  print("Digits: ", digits)
  print("Alphabets: ", alphabets)
  print("Spaces: ", spaces)
```

```python
counts()
```

```python
# Task 3

# Write a program that asks for a sentence and counts:
# Digits
# Alphabets
# Spaces

def counts_ai():
  sentence = input("Enter a sentence: ")
  num_digits = sum(1 for char in sentence if char.isdigit())
  num_alphabets = sum(1 for char in sentence if char.isalpha())
  num_spaces = sum(1 for char in sentence if char == " " )
  print("Digits: ", num_digits)
  print("Alphabets: ",num_alphabets)
  print("Spaces: ",num_spaces)
```

```python
counts_ai()
```
