# 1.Write a program to print the sum of numbers from 1 to 100




```python
total_sum = 0
for i in range(1, 101):
    total_sum += i
print(f"The sum of numbers from 1 to 100 is: {total_sum}")
```

# 2. Ask the user for a number → print its factorial.



```python
# 2. Ask the user for a number → print its factorial.
fact = int(input("Enter a number: "))
factorial = 1
for i in range(1, fact + 1):
    factorial *= i
print(f"The factorial of {fact} is: {factorial}")
```

## 3. Create a program that asks for a password and gives the user only 3 attempts

```python
# Create a program that asks for a password and gives the user only 3 attempts
password = "smit123"
attempts = 0
while attempts < 3:
    entered_pass = input("Enter the password: ")
    if entered_pass == password:
        print("Access granted!")
        break
    else:
        print("Access denied. Try again.")
        attempts += 1
```
