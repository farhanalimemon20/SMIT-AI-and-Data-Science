```python
print("Welcome to Simple Calculator")

print("""
Choose an operation:
+  for Addition
-  for Subtraction
*  for Multiplication
/  for Division
%  for Modulus
// for Floor Division
** for Power
age to Calculate Age
BMI to calculate your BMI
""")

op = input("Enter operation: ")

if op in ["+", "-", "*", "/", "%", "//", "**"]:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

    if op == "+":
        print("Result =", num1 + num2)
    elif op == "-":
        print("Result =", num1 - num2)
    elif op == "*":
        print("Result =", num1 * num2)
    elif op == "/":
        if num2 != 0:
            print("Result =", num1 / num2)
        else:
            print("Error: Cannot divide by zero")
    elif op == "%":
        print("Result =", num1 % num2)
    elif op == "//":
        print("Result =", num1 // num2)
    elif op == "**":
        print("Result =", num1 ** num2)

elif op == "age":
    birth_year = int(input("Enter your birth year: "))
    current_year = 2025
    age = current_year - birth_year
    print("Your age is:", age)
elif op == "BMI":
    height =(float(input("Enter your height in meters")))
    weight = (int(input("Enter your weight in kg")))
    print("Your BMI is :",weight/(height**2))
else:
    print("Invalid operation")

```

# Write a program that asks 5 subject marks and prints highest and lowest.

```python
marks = []
for i in range(5):
  mark = int(input(f"Enter marks for subject No:{i+1} = "))
  marks.append(mark)
highest_mark = max(marks)
lowest_mark = min(marks)
print("Highest mark =", highest_mark)
print("Lowest mark =", lowest_mark)
```

```python
marks ={}
subjects = ["English","Urdu","Maths","Science","Islamiat"]
for subject in subjects:
  mark = int(input(f"Enter Your Marks for {subject} ="))
  marks[subject] = mark
highest_mark = max(marks.values())
lowest_mark = min(marks.values())
print("Highest mark =", highest_mark)
print("Lowest mark =", lowest_mark)


```
