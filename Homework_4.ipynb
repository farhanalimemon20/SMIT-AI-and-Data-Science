# 4.Use sets to store unique numbers entered by a user. Print them sorted.

```python
# Program to store unique numbers using sets and print them sorted

# Create an empty set
unique_numbers = set()

# Ask user how many numbers they want to enter
n = int(input("How many numbers do you want to enter? "))

# Take input from user
for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    unique_numbers.add(num)   # add ensures uniqueness

# Print sorted numbers
print("Unique numbers in sorted order:", sorted(unique_numbers))
```

```python
unique_numbers = set()
num_inputs = int(input("How many numbers do you want to enter? "))

for i in range(num_inputs):
    try:
        num = int(input(f"Enter number {i+1}: "))
        unique_numbers.add(num)
    except ValueError:
        print("Invalid input. Please enter an integer.")

sorted_unique_numbers = sorted(list(unique_numbers))
print("Sorted unique numbers:", sorted_unique_numbers)
```
